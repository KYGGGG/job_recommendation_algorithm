```python
# bq4_1a
num = []
for num1 in range(0, len(CER)):
    for num2 in range(0,len(x_train)):
        if CER['BE'][num1] in x_train['bq4_1a'][num2]:
            x_train['bq4_1a'][num2] = CER['AF'][num1]
            num.append(num2)
a = dict.fromkeys(num)
a = list(a)
b = list(range(0, len(x_train)))
for i in range(0, len(a)):
    b.remove(a[i])
x_train['bq4_1a'][b] = '없음'
```

```python
# bq4_1b
num = []
for num1 in range(0, len(CER)):
    for num2 in range(0,len(x_train)):
        if CER['BE'][num1] in x_train['bq4_1b'][num2]:
            x_train['bq4_1b'][num2] = CER['AF'][num1]
            num.append(num2)
a = dict.fromkeys(num)
a = list(a)
b = list(range(0, len(x_train)))
for i in range(0, len(a)):
    b.remove(a[i])
x_train['bq4_1b'][b] = '없음'
```

```python
# bq4_1c
num = []
for num1 in range(0, len(CER)):
    for num2 in range(0,len(x_train)):
        if CER['BE'][num1] in x_train['bq4_1c'][num2]:
            x_train['bq4_1c'][num2] = CER['AF'][num1]
            num.append(num2)
a = dict.fromkeys(num)
a = list(a)
b = list(range(0, len(x_train)))
for i in range(0, len(a)):
    b.remove(a[i])
x_train['bq4_1c'][b] = '없음'
```

```python
LEC = pd.read_csv('C:/Users/zihyu/DACON/EX/LEC.csv')
# bq38_1
x_train['bq38_1_1'] = 0
x_train['bq38_1'] = x_train['bq38_1'].fillna('없음')

for BE in LEC['BE'] :
    for str in x_train['bq38_1'] : 
        if BE in str:
            x_train['bq38_1'].replace(str, LEC['AF'].values[LEC['BE'] == BE][0], inplace = True)

for BE in LEC['BE'] :
    x_train['bq38_1_1'][x_train.index[x_train['bq38_1'] == BE].tolist()] = 1
    
for i in range(0,len(x_train['bq38_1_1']),1) :
    if x_train['bq38_1_1'][i] == 0:
            x_train['bq38_1'].replace(x_train['bq38_1'][i], "없음", inplace = True)

x_train.drop(columns = 'bq38_1_1', inplace = True)
```

```python
BE_J = pd.read_csv('C:/Users/zihyu/DACON/EX/BE_J.csv')
# bq32 이전 작업
x_train['bq32_1'] = 0
x_train['bq32'] = x_train['bq32'].fillna('없음')

for BE in BE_J['BE'] :
    for str in x_train['bq32'] : 
        if BE in str:
            x_train['bq32'].replace(str, BE_J['AF'].values[BE_J['BE'] == BE][0], inplace = True)

for BE in BE_J['BE'] :
    x_train['bq32_1'][x_train.index[x_train['bq32'] == BE].tolist()] = 1
    
for i in range(0,len(x_train['bq32_1']),1) :
    if x_train['bq32_1'][i] == 0:
            x_train['bq32'].replace(x_train['bq32'][i], "없음", inplace = True)

x_train.drop(columns = 'bq32_1', inplace = True)
```

```python
#연봉 평균 대치
idx = x_train['bq41'][x_train['bq41'].isnull()].index

idx_1 = x_train['bq41'][x_train['bq41'].isnull() == False].index

for i in idx_1 :
    x_train['bq41'][i] = int(x_train['bq41'][i])
    
for i in idx :
    x_train['bq41'][i] = int(np.mean(x_train['bq41'][x_train['knowcode'] == x_train['knowcode'][i]]))
```
