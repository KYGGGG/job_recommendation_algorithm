```python
import pandas as pd
import numpy as np
from scipy.stats import mode    #최빈값

from sklearn.model_selection import train_test_split, GridSearchCV #스플릿
from sklearn.ensemble import RandomForestClassifier #랜덤포레스트
from sklearn.metrics import accuracy_score #스코어
from sklearn.preprocessing import StandardScaler #표준화

```

```python
df = pd.read_csv("C:/Users/zihyu/DACON/EX/END.csv", encoding = 'cp949')
```

```python
df.drop(columns = ['Unnamed: 0'])
```

```python
df.info()
```

```python
categorical = [var for var in df.columns if df[var].dtype=='O']
categorical
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for ca in categorical :
    le.fit(df[ca])
    df[ca] = le.transform(df[ca])
```

```python
x = df.drop(['knowcode', 'Unnamed: 0','bq32'], axis=1, )
y = df['knowcode']
```

```python
x
```

```python
y
```

```python
X_train, X_val, y_train, y_val = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape, X_val.shape
```

```python
clf = RandomForestClassifier(n_estimators=300, max_depth =50, random_state=0)
clf.fit(X_train,y_train)

predict1 = clf.predict(X_val)
print(accuracy_score(y_val,predict1))
```
